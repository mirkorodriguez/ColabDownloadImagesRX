```
# Copying final files
import os
import shutil
import glob

import pandas as pd

PATH = "/content/drive/My Drive/courses/DL/Download/"

## ----------------------------
## PROCESANDO IMAGENES NORMALES
## ----------------------------

src_dir = "".join([PATH,"ChestXRay/chest_xray/train/NORMAL/"])
dst_dir = "".join([PATH,"dataset/normal/"])
total_to_copy = 220
num_files = 0

files = glob.glob(src_dir+"*.jpeg")

# Copiando archivos COVID-19
for i in range(0,total_to_copy):
    filename = files[i]
    num_files+=1
    # print(filename)
    # Copy files
    shutil.copy(os.path.join(src_dir, filename), os.path.join(dst_dir, 'NORMAL-' + str(num_files).zfill(3) + '.jpg'))

print("Num de imgs procesadas:", num_files)
```
