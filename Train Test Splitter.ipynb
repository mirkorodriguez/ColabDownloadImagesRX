```
# Copying final files
import os
import shutil
import glob

import pandas as pd

PATH = "/content/drive/My Drive/courses/DL/Download/dataset/"

## -------------------
## PROCESANDO IMAGENES
## -------------------
total_files_covid19 = glob.glob(PATH+"covid19/*.jpg")
total_files_normal = glob.glob(PATH+"normal/*.jpg")

TRAIN_SIZE = 0.8

# COVID19
src_covid19_dir = "".join([PATH,"covid19/"])
dst_covid19_dir_train = "".join([PATH,"train/covid19/"])
dst_covid19_dir_test = "".join([PATH,"test/covid19/"])

total_to_copy_covid19_train = round(TRAIN_SIZE*len(total_files_covid19))

num_files = 0
for i in range(0,total_to_copy_covid19_train):
    filename = total_files_covid19[i].split("/")[-1]
    num_files+=1
    # Copy files
    shutil.copy(os.path.join(src_covid19_dir, filename), os.path.join(dst_covid19_dir_train, filename))
print("Num de imgs procesadas COVID19-TRAIN:", num_files)

num_files = 0
for j in range(total_to_copy_covid19_train,len(total_files_covid19)):
    filename = total_files_covid19[j].split("/")[-1]
    num_files+=1
    # Copy files
    shutil.copy(os.path.join(src_covid19_dir, filename), os.path.join(dst_covid19_dir_test, filename))
print("Num de imgs procesadas COVID19-TEST:", num_files)

# NORMAL
src_normal_dir = "".join([PATH,"normal/"])
dst_normal_dir_train = "".join([PATH,"train/normal/"])
dst_normal_dir_test = "".join([PATH,"test/normal/"])

total_to_copy_normal_train = round(TRAIN_SIZE*len(total_files_normal))

num_files = 0
for i in range(0,total_to_copy_normal_train):
    filename = total_files_normal[i].split("/")[-1]
    num_files+=1
    # Copy files
    shutil.copy(os.path.join(src_normal_dir, filename), os.path.join(dst_normal_dir_train, filename))
print("Num de imgs procesadas NORMAL-TRAIN:", num_files)

num_files = 0
for j in range(total_to_copy_normal_train,len(total_files_normal)):
    filename = total_files_normal[j].split("/")[-1]
    num_files+=1
    # Copy files
    shutil.copy(os.path.join(src_normal_dir, filename), os.path.join(dst_normal_dir_test, filename))
print("Num de imgs procesadas NORMAL-TEST:", num_files)
```
