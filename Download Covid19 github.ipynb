```
import os
import shutil
import glob

import pandas as pd

# Funciones
ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg','PNG','JPG','JPEG'])
PATH = "/content/drive/My Drive/courses/DL/Download/"

def allowed_file(filename):
    return '.' in filename and filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS

## ------------------------------
## PROCESANDO IMAGENES DE COVID19
## ------------------------------

# Leyendo metadata
metadata = pd.read_csv("".join([PATH,"Covid19/covid-chestxray-dataset/metadata.csv"]))
print(metadata.head())

files = metadata[["filename","finding"]]
print(files.head())
print(files.shape)

src_dir = "".join([PATH,"Covid19/covid-chestxray-dataset/images/"])
dst_dir = "".join([PATH,"dataset/covid19/"])
num_files = 0

# Copiando archivos COVID-19
for i in files.index:
    filename = files['filename'][i]
    finding = files['finding'][i]

    if(allowed_file(filename) and (finding == 'COVID-19')):
        num_files+=1
        # print(filename, "\t" ,finding)
        # Copy files
        shutil.copy(os.path.join(src_dir, filename), os.path.join(dst_dir, 'COVID19-' + str(num_files).zfill(3) + '.jpg'))

print("Num de imgs procesadas:", num_files)
```
